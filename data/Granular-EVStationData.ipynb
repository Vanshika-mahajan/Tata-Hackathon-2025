```python
df_stations=pd.read_json("ev_charging_stations_granular.json")
```

```python
df_stations
```

```python
df_stations.rename(columns={
    'lattitude': 'latitude',
    'longitude': 'longitude',
    'name': 'station_name',
    'address': 'address',
    'city': 'city',
    'state': 'state',
    'pricing': 'pricing',
    'availability': 'availability_status', # Renaming for clarity
    'type': 'charger_type_code' # Assuming 'type' is a code for charger type
}, inplace=True)
```

```python
df_stations
```

```python
columns_to_drop = ['contact', 'isValidated', 'issue', 'notes', 'phones', 'stationid']
df_stations.drop(columns=columns_to_drop, errors='ignore', inplace=True)
```

```python
# Convert to numeric, coercing errors (like 'NA' or empty strings) to NaN
df_stations['latitude'] = pd.to_numeric(df_stations['latitude'], errors='coerce')
df_stations['longitude'] = pd.to_numeric(df_stations['longitude'], errors='coerce')
```

```python
# Drop rows where Latitude or Longitude are NaN (after coercion)
df_stations.dropna(subset=['latitude', 'longitude'], inplace=True)
```

```python
#Handle Duplicates
# Identify unique stations by Lat/Lon and name
df_stations.drop_duplicates(subset=['latitude', 'longitude', 'station_name'], inplace=True)
```

```python
# Further interpretation of 'charger_type_code' (if possible from documentation)
# If 'type' is always '0' and represents AC chargers, you can create a readable column:
df_stations['charger_type_details'] = df_stations['charger_type_code'].astype(str).replace({'0': 'AC Charger (Slow)'})
# If you find documentation that '1' means DC Fast, you'd add:
# .replace({'0': 'AC Charger (Slow)', '1': 'DC Fast Charger'})
# For now, let's assume '0' is some form of standard AC charger.
```
